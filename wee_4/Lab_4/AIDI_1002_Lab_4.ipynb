```python
# Load data
df = pd.read_csv('CAIT 2.0 Country Clean Technology Data-csv.csv')
df.head(10)
```

#### Description of the columns

- Country: Country Name
- Year: year
- Cumulaive Installed Solar PV Capacity (MW): Cumulative amount energy can be produced by the solar system
- Annual Additions of Installed Solar PV Capacity (MW): the additional amount of energy that can be produced by solar system compare to last year
- Annual Manufacturing capacity of Sillicon (Metric Tons): amount of silicon can be produced in a year
- Annual Manufacturing Capacity of PV Cells (MW): the number of photovoltaic cells that can be produced in a year based on the amount of energy they can produce
- Annual Manufacturing Capacity of PV modules (MW): number of photovoltaic modules that can be manufactured in a year based on the amount of energy the would produce
- Annual Manufacturing of Wind Turbines (MW): the number of wind turbines that can be producecd in a year based on the amount of energy they output
- Total Solar PV Jobs: total number of solar photovoltaic jobs
- Total Wind Jobs: total number of wind energy jobs
- Total Manufacturing Jobs: total number of turbine manufacturing jobs
